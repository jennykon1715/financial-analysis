```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

#下載股票數據，同時看多支股票
tickers = ['2330.TW', '2454.TW', '3711.TW']
today = datetime.now().strftime('%Y-%m-%d')
data = yf.download(tickers,start='2023-01-01', end=today)


#獲取收盤價數據
df_close = data['Close']

#創建一個圖表和三個子圖
plt.figure(figsize=(14, 7))
for i, t in enumerate(tickers, 1):
    plt.subplot(3, 1, i)  # 3行1列，當前位置i
    plt.plot(df_close.index, df_close[t], label=f'{t} Closing Price')
    plt.legend()
    plt.tight_layout()


plt.show()



#plt.figure(figsize=(8,4))
#for t in tickers:
 # plt.plot(df_close.index, df_close[t], label=t)
  #plt.legend()
  #plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

#下載股票數據，同時看多支股票
tickers = ['2330.TW', '2454.TW', '3711.TW']
today = datetime.now().strftime('%Y-%m-%d')
data = yf.download(tickers,start='2023-01-01', end=today)

#獲取收盤價數據
df_close = data['Close']

# 獲取每個股票的名稱
stock_names = {}
for ticker in tickers:
    info = yf.Ticker(ticker).info
    stock_names[ticker] = info.get('shortName', '')  # 使用 'shortName' 作為股票名稱，如果沒有找到就使用空字符串

# 創建一個圖表
plt.figure(figsize=(14, 7))

# 為每個股票繪製收盤價趨勢線，並添加標籤包含股票代碼和名稱
for t in tickers:
    label = f'{t} / {stock_names[t]}'
    plt.plot(df_close.index, df_close[t], label=label)

# 設置圖例
plt.legend(title='Closing Price')

# 設置標題和軸標籤
plt.title('Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```
